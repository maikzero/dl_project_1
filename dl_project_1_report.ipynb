# Initial model
This is our initial model with no augmentations to the data. This results in a Kaggle accuracy of around 7.8%. The model was unable to handle any distortions and as a result was unable to recognize any of the images. This is where we realized that we needed to make the model more robust.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV files from the 'csv' folder
train_data = pd.read_csv('csv_initial/train.csv')
val_acc_data = pd.read_csv('csv_initial/val_acc.csv')
val_loss_data = pd.read_csv('csv_initial/val_loss.csv')

# Select only the first two columns (Step and Value) for each dataset
train_data = train_data.iloc[:, :2]
val_acc_data = val_acc_data.iloc[:, :2]
val_loss_data = val_loss_data.iloc[:, :2]

# Rename columns for clarity (optional)
train_data.columns = ['Step', 'Value']
val_acc_data.columns = ['Step', 'Value']
val_loss_data.columns = ['Step', 'Value']

# Display the first few rows of each dataframe to confirm
print("Training Data:")
print(train_data.head())

print("\nValidation Accuracy Data:")
print(val_acc_data.head())

print("\nValidation Loss Data:")
print(val_loss_data.head())

# Plotting the data
plt.figure(figsize=(14, 6))

# Plot Training Data
plt.subplot(1, 3, 1)
sns.lineplot(x='Step', y='Value', data=train_data)
plt.title('Training Data')
plt.xlabel('Step')
plt.ylabel('Value')

# Plot Validation Accuracy
plt.subplot(1, 3, 2)
sns.lineplot(x='Step', y='Value', data=val_acc_data)
plt.title('Validation Accuracy')
plt.xlabel('Step')
plt.ylabel('Accuracy')

# Plot Validation Loss
plt.subplot(1, 3, 3)
sns.lineplot(x='Step', y='Value', data=val_loss_data)
plt.title('Validation Loss')
plt.xlabel('Step')
plt.ylabel('Loss')

plt.tight_layout()
plt.show()
```
